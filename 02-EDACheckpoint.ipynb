# Research Question

How do FAANG companies differ in their daily return performance and volatility across market regimes defined by VIX levels, and which firms exhibit greater resilience during periods of elevated market stress?

Market conditions are classified into three regimes based on the CBOE Volatility Index (VIX): low volatility (VIX < 20), representing stable market conditions with investor confidence; moderate uncertainty (20 ≤ VIX < 30), indicating elevated investor concern; and high stress (VIX ≥ 30), reflecting acute market turbulence and risk aversion. These thresholds are consistent with established practitioner conventions and prior academic literature (Whaley, 2009).

Resilience is defined as a firm’s ability to maintain relatively stable mean daily returns and limit increases in return standard deviation during high-stress regimes (VIX ≥ 30) compared to low-stress conditions (VIX < 20).

## Background and Prior Work

FAANG refers to five major U.S. technology companies: Meta (Facebook), Apple, Amazon, Netflix, and Google (Alphabet). These firms are among the largest publicly traded companies in the world and collectively account for a significant share of the total market capitalization of the S&P 500. Because of their size and market influence, movements in FAANG stock prices can meaningfully affect broader market performance.

Although these firms are frequently grouped together, their underlying business models differ substantially. Apple relies heavily on hardware sales and a premium consumer ecosystem, Google generates most of its revenue through digital advertising, Netflix depends almost entirely on subscription-based streaming services, and Amazon derives significant income from e-commerce and cloud computing through AWS. These structural differences are consequential: prior research comparing large-cap technology firms demonstrates that business model heterogeneity produces meaningfully different return distributions, even among firms that appear similar in market capitalization or sector classification (Hasanhodzic & Lo, 2007). In particular, firms with recurring revenue models (e.g., subscriptions, cloud contracts) tend to exhibit lower return volatility than firms reliant on discretionary consumer spending or advertising budgets, which are more sensitive to macroeconomic cycles.

Prior research suggests that differences in firm structure and revenue exposure can influence how companies respond to macroeconomic shocks and investor sentiment (Fama & French, 1992). Firms with revenue streams tied closely to consumer demand or discretionary spending may experience greater price fluctuations during periods of economic uncertainty, while firms with diversified or recurring revenue models may exhibit more stable performance.

Volatility, defined as the degree to which a stock’s price fluctuates over time, is widely used as a measure of financial risk. Research shows that volatility is not random but instead clusters over time and responds strongly to market shocks (Engle & Patton, 2001). During periods of market stress, negative information tends to produce larger increases in volatility than positive news of the same magnitude — a phenomenon known as the leverage effect. These dynamics suggest that market-wide uncertainty captured by the VIX may affect individual FAANG firms differently depending on their business model risk exposure and investor base.

Much of the previous research examines volatility using aggregate measures such as full-sample standard deviation or beta relative to the market. While these methods provide useful information about long-run risk, they often pool together periods of both stable and turbulent market conditions, causing important differences in firm behavior across market environments to be obscured. Existing regime-based volatility studies (e.g., Hamilton, 1989; Ang & Bekaert, 2002) typically focus on index-level behavior or apply hidden Markov models, and do not directly compare individual firm responses within the same sector across pre-specified, interpretable VIX thresholds. This project addresses that gap by using a transparent, rule-based regime classification that is accessible to practitioners and allows direct firm-to-firm comparison.

### Netflix (NFLX) Daily Stock Prices

This dataset contains daily historical stock market data for Netflix (NFLX) from May 2002 through mid-2020, spanning over 4,500 trading days. Each row represents one trading day — weekends and market holidays are excluded. All price columns are measured in **U.S. dollars (USD) per share**.

- **Open** is the price at which Netflix stock first traded when the market opened that day.
- **High** and **Low** are the highest and lowest prices reached during that trading session, giving a sense of intraday price range and turbulence.
- **Close** is the final traded price at the end of the session.
- **Adj Close** is the closing price adjusted for corporate actions, most importantly stock splits. Netflix underwent a 7-for-1 stock split in 2015, which is why early-era prices appear very small (around $1) — they have been retroactively scaled down for consistency. For our analysis, **Adj Close is the correct variable to use** when computing returns, as it produces a consistent and comparable price series across the full date range. From it we derive the **daily log return** (`log(Adj Close_t / Adj Close_{t-1})`), which measures how much the stock moved each day. Volatility is then quantified as the rolling standard deviation of those log returns — higher values signal greater day-to-day price swings and therefore higher risk.
- **Volume** is the number of shares traded during the session (unit: shares). Unusually high volume days often coincide with earnings releases, major company news, or broad market stress events, and can serve as a secondary marker of market turbulence.

There are a few notable concerns with this dataset. First, it contains only Netflix-specific price data and no external market variables such as S&P 500 returns or VIX levels. To classify market conditions as "favorable" or "unfavorable" for our analysis, we will need to merge in a separate benchmark dataset. Second, Netflix's business model changed dramatically across the dataset's time range — from a DVD-by-mail service (2002–2007), to a domestic streaming platform (2007–2016), to a global content producer (2016–present). The drivers of its stock volatility differ substantially across these eras, which is a meaningful confound when interpreting results across the full period. Third, the earliest records (2002–2004) reflect a newly public, thinly traded small-cap stock with speculative behavior, producing unusually high volatility that may not be representative of the mature company's risk profile. These early years should be interpreted with caution.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# ── Paths ──────────────────────────────────────────────────────────────────────
RAW_PATH      = Path("data/00-raw/Netflix.csv")
INTERIM_DIR   = Path("data/01-interim")
PROCESSED_DIR = Path("data/02-processed")
INTERIM_DIR.mkdir(parents=True, exist_ok=True)
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

# ── 1. Load ────────────────────────────────────────────────────────────────────
df_raw = pd.read_csv(RAW_PATH)
print("Raw shape:", df_raw.shape)
display(df_raw.head())

# ── 2. Type enforcement ────────────────────────────────────────────────────────
df = df_raw.copy()
df.columns = [c.strip() for c in df.columns]
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df = df.sort_values("Date").reset_index(drop=True)

num_cols = ["Open", "High", "Low", "Close", "Adj Close", "Volume"]
for c in num_cols:
    df[c] = pd.to_numeric(df[c], errors="coerce")

print("\nData types:")
print(df.dtypes)
print("\nDataset size (rows, cols):", df.shape)

# ── 3. Missingness ─────────────────────────────────────────────────────────────
missing_summary = pd.DataFrame({
    "missing_count": df.isna().sum(),
    "missing_pct":   (df.isna().sum() / len(df) * 100).round(4)
}).sort_values("missing_count", ascending=False)

print("\nMissingness summary:")
display(missing_summary)

# Check whether missingness clusters in any particular year
df["missing_any"] = df.isna().any(axis=1)
missing_by_year = df.groupby(df["Date"].dt.year)["missing_any"].mean()
print("\nMissing rate by year (only years with missing data shown):")
display(missing_by_year[missing_by_year > 0])

# ── 4. Tidy check: OHLC internal consistency ──────────────────────────────────
suspicious = df[
    (df["Low"]   > df["High"]) |
    (df["Open"]  < df["Low"])  | (df["Open"]  > df["High"]) |
    (df["Close"] < df["Low"])  | (df["Close"] > df["High"])
]
print(f"\nSuspicious OHLC rows (internal inconsistencies): {len(suspicious)}")
if len(suspicious) > 0:
    display(suspicious)

# ── 5. Derived columns & outlier flagging ─────────────────────────────────────
df["Daily_Log_Return"] = np.log(df["Adj Close"] / df["Adj Close"].shift(1))
df["Price_Range"]      = df["High"] - df["Low"]

# Extreme return days (beyond 1st/99th percentile)
ret_lo, ret_hi = df["Daily_Log_Return"].quantile([0.01, 0.99])
extreme_ret = df[(df["Daily_Log_Return"] <= ret_lo) | (df["Daily_Log_Return"] >= ret_hi)]
print(f"\nExtreme return thresholds — 1st pct: {ret_lo:.4f}, 99th pct: {ret_hi:.4f}")
print(f"Extreme return days flagged: {len(extreme_ret)}")
display(extreme_ret[["Date", "Daily_Log_Return", "Adj Close", "Volume"]].head(10))

# Extreme volume days (above 99th percentile)
vol_thresh = df["Volume"].quantile(0.99)
high_vol   = df[df["Volume"] >= vol_thresh]
print(f"\nHigh volume threshold (99th pct): {vol_thresh:,.0f} shares")
print(f"High volume days flagged: {len(high_vol)}")
display(high_vol[["Date", "Volume", "Adj Close", "Daily_Log_Return"]].head(10))

# Note: extreme days in financial data represent real market events (earnings,
# crashes, major announcements) — they are not errors and should NOT be removed,
# as they are precisely the observations most relevant to our volatility analysis.

# ── 6. Summary statistics ──────────────────────────────────────────────────────
print("\nSummary statistics — price & volume columns:")
display(df[num_cols].describe())

print("\nSummary statistics — daily log returns:")
display(df[["Daily_Log_Return"]].describe())

# ── 7. Visualizations ─────────────────────────────────────────────────────────
fig, axes = plt.subplots(3, 1, figsize=(14, 12))

axes[0].plot(df["Date"], df["Adj Close"], color="royalblue", linewidth=0.8)
axes[0].set_title("Netflix Adjusted Closing Price Over Time")
axes[0].set_ylabel("Adj Close (USD)")
axes[0].grid(alpha=0.3)

axes[1].plot(df["Date"], df["Daily_Log_Return"], color="darkorange", linewidth=0.5, alpha=0.8)
axes[1].axhline(0, color="black", linewidth=0.8, linestyle="--")
axes[1].set_title("Netflix Daily Log Returns")
axes[1].set_ylabel("Log Return")
axes[1].grid(alpha=0.3)

rolling_vol = df["Daily_Log_Return"].rolling(window=30).std()
axes[2].plot(df["Date"], rolling_vol, color="firebrick", linewidth=0.8)
axes[2].set_title("Netflix 30-Day Rolling Volatility (Std of Log Returns)")
axes[2].set_ylabel("Volatility")
axes[2].grid(alpha=0.3)

plt.tight_layout()
plt.show()

# ── 8. Clean & save ───────────────────────────────────────────────────────────
before = len(df)

# Drop rows missing any core price column — these are unusable for return calculations.
# We use dropna on the price columns only (not Volume, which has occasional gaps
# that don't affect return computation). This is a conservative choice that drops
# very few rows while ensuring no NaN values propagate into our return series.
df_clean = df.dropna(subset=["Date"] + num_cols).copy()
df_clean = df_clean.drop(columns=["missing_any"], errors="ignore")

after = len(df_clean)
print(f"\nRows before cleaning: {before} | After: {after} | Dropped: {before - after}")
print(f"Date range: {df_clean['Date'].min().date()} → {df_clean['Date'].max().date()}")

df.to_csv(INTERIM_DIR / "netflix_interim.csv", index=False)
df_clean.to_csv(PROCESSED_DIR / "netflix_processed.csv", index=False)
print("Files saved.")

display(df_clean.head())
df_clean.info()
```

### Google Daily Stock Prices

This dataset contains daily historical stock market data for Google (Alphabet Inc.). Each row represents one trading day (so weekends and market holidays are excluded). The dataset includes standard market variables used in financial analysis: Open, High, Low, Close, Adjusted Close (Adj Close), and Volume, along with the Date.

All price-related variables are measured in U.S. dollars (USD) per share.

- **Open** is the first traded price when the market opens that day.
- **High** and **Low** are the highest and lowest traded prices during the trading session.
- **Close** is the last traded price at market close.
- **Adj Close** is the closing price adjusted for corporate actions (primarily stock splits). For Google's data in this dataset (2004–2020), Close and Adj Close are identical throughout, since the 20-for-1 stock split Google performed occurred in 2022 till after this dataset ends. For return calculations over time, Adj Close is still the recommended column to use for consistency.
- **Volume** is the number of shares traded that day (unit: shares). Higher volume often indicates increased investor attention, news/earnings events, or higher volatility. Google's daily volume tends to be much lower than other FAANG stocks in raw share count because its per-share price is significantly higher (reaching ~$1,700 by 2020).

**Major concerns/limitations:** This dataset only includes market outcomes (prices/volume) and does not include explanatory variables like interest rates, inflation, sector indices, earnings surprises, or macroeconomic indicators. This means any "why" interpretation is limited unless additional datasets are incorporated. Additionally, the dataset begins at Google's IPO in August 2004, meaning it does not capture any pre-public company history, and the IPO period itself exhibits unusually high volatility typical of newly listed stocks. The dataset ends in September 2020, so it does not reflect more recent market conditions. Finally, "outliers" in finance are frequently real events (earnings, crashes, major announcements) rather than data errors, so outlier handling must be done carefully.

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Paths
raw_path = "data/00-raw/Google.csv"
interim_path = "data/01-interim/google_interim.csv"
processed_path = "data/02-processed/google_clean.csv"

# Load the dataset
df = pd.read_csv(raw_path, parse_dates=['Date'])
print("Dataset loaded successfully.\n")

# Check first few rows and columns
print("First 5 rows of the dataset:")
display(df.head())
print("\nColumns and data types:")
print(df.dtypes)

# Check size of dataset
print(f"\nDataset shape: {df.shape} (rows, columns)")

# Check missing values
missing_counts = df.isnull().sum()
print("\nMissing values per column:")
print(missing_counts)

# Identify outliers (z-score > 3)
numeric_cols = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']
for col in numeric_cols:
    df[f'{col}_zscore'] = (df[col] - df[col].mean()) / df[col].std()

# Flag extreme outliers
outliers = df[(df[[f'{col}_zscore' for col in numeric_cols]]).abs() > 3].dropna(how='all')
print(f"\nPotential outliers flagged: {len(outliers)}")
display(outliers)

# Clean dataset: drop rows with missing Adj Close
df_clean = df.dropna(subset=['Adj Close']).reset_index(drop=True)

# Drop z-score columns
df_clean = df_clean.drop(columns=[f'{col}_zscore' for col in numeric_cols])

# Save intermediate and processed datasets
df_clean.to_csv(interim_path, index=False)
print(f"Intermediate cleaned dataset saved to {interim_path}")

df_clean.to_csv(processed_path, index=False)
print(f"Processed dataset saved to {processed_path}")

# Summary after cleaning
print("\nSummary after cleaning:")
print(df_clean.info())
print(df_clean.describe())
```

#### T-Test Analysis: FAANG Performance Across Market Regimes

To formally test whether FAANG companies behave differently during periods of market stress versus calm, we run independent-samples t-tests comparing two groups of trading days. We classify days where the VIX condition is **Low** as **Good** market conditions, and days where VIX is **High** as **Bad** market conditions. We run this test twice — once for **daily returns**, and once for **annualized volatility** — to capture both the direction and the risk dimension of performance.

Since we run five simultaneous tests (one per company) for each analysis, we apply a **Bonferroni correction** to control the family-wise Type I error rate, multiplying each raw p-value by 5.

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

faang_df = pd.read_csv('data/02-processed/faang_processed.csv')
faang_df = faang_df.dropna(subset=['Returns', 'Volatility'])
faang_df['Market_Regime'] = faang_df['vix_condition'].apply(
    lambda x: 'Good' if x in ['Low', 'Moderate'] else 'Bad'
)

companies = sorted(faang_df['Company'].unique())
COLORS = {'Good': '#2ecc71', 'Bad': '#e74c3c'}
COMPANY_COLORS = {
    'Amazon': '#FF9900', 'Apple': '#555555', 'Facebook': '#1877F2',
    'Google': '#4285F4', 'Netflix': '#E50914'
}
good_patch = mpatches.Patch(color=COLORS['Good'], alpha=0.8, label='Good Market (VIX: Low)')
bad_patch  = mpatches.Patch(color=COLORS['Bad'],  alpha=0.8, label='Bad Market (VIX: High)')
```

##### Part 1: Daily Returns — Good vs. Bad Market Conditions

We test whether mean daily returns differ significantly between good and bad market regimes for each company. The notched box plots below show the full return distributions — non-overlapping notches indicate statistically significant median differences.

- **H₀**: Mean daily returns are equal across market regimes
- **H₁**: Mean daily returns are lower during bad market conditions

```python
fig, axes = plt.subplots(1, 5, figsize=(18, 6), sharey=True)
fig.suptitle(
    'Daily Returns by Market Regime Across FAANG Companies\n'
    '(Good = Low/Moderate VIX  |  Bad = Elevated/High/Extreme VIX)',
    fontsize=14, fontweight='bold', y=1.02)

for ax, comp in zip(axes, companies):
    sub = faang_df[faang_df['Company'] == comp]
    data_good = sub[sub['Market_Regime'] == 'Good']['Returns'] * 100
    data_bad  = sub[sub['Market_Regime'] == 'Bad']['Returns']  * 100
    bp = ax.boxplot([data_good, data_bad],
                    patch_artist=True, notch=True, widths=0.5,
                    medianprops=dict(color='white', linewidth=2),
                    flierprops=dict(marker='o', markersize=2, alpha=0.3))
    bp['boxes'][0].set_facecolor(COLORS['Good']); bp['boxes'][0].set_alpha(0.8)
    bp['boxes'][1].set_facecolor(COLORS['Bad']);  bp['boxes'][1].set_alpha(0.8)
    ax.set_title(comp, fontsize=12, fontweight='bold', color=COMPANY_COLORS[comp])
    ax.set_xticks([1, 2])
    ax.set_xticklabels(['Good\nMarket', 'Bad\nMarket'], fontsize=9)
    ax.axhline(0, color='gray', linestyle='--', linewidth=0.8, alpha=0.6)

axes[0].set_ylabel('Daily Return (%)', fontsize=11)
fig.legend(handles=[good_patch, bad_patch], loc='lower center', ncol=2,
           fontsize=10, bbox_to_anchor=(0.5, -0.05), frameon=True)
plt.tight_layout()
plt.show()
```

```python
return_results = []
for comp in companies:
    sub = faang_df[faang_df['Company'] == comp]
    good = sub[sub['Market_Regime'] == 'Good']['Returns'].dropna()
    bad  = sub[sub['Market_Regime'] == 'Bad']['Returns'].dropna()
    t, p = stats.ttest_ind(good, bad)
    diff = (good.mean() - bad.mean()) * 100
    se = np.sqrt(good.std()**2/len(good) + bad.std()**2/len(bad)) * 100
    ci_lo, ci_hi = diff - 1.96*se, diff + 1.96*se
    p_bonf = min(p * 5, 1.0)
    return_results.append({
        'Company':         comp,
        'Mean Good (%)':   round(good.mean()*100, 3),
        'Mean Bad (%)':    round(bad.mean()*100, 3),
        'Difference (%)':  round(diff, 3),
        '95% CI':          '[{:.3f}, {:.3f}]'.format(ci_lo, ci_hi),
        't-stat':          round(t, 3),
        'p (Bonferroni)':  '<0.001' if p_bonf < 0.001 else round(p_bonf, 4)
    })

return_df = pd.DataFrame(return_results)
display(return_df.style.set_caption('Table 1: T-Test Results — Daily Returns (Good vs. Bad Market)'))
```

```python
fig, ax = plt.subplots(figsize=(11, 5))
bar_colors = [COMPANY_COLORS[c] for c in return_df['Company']]
diffs = return_df['Difference (%)'].values
ci_lo_vals = np.array([float(s.strip('[]').split(',')[0]) for s in return_df['95% CI']])
ci_hi_vals = np.array([float(s.strip('[]').split(',')[1]) for s in return_df['95% CI']])

ax.barh(return_df['Company'], diffs, color=bar_colors, alpha=0.85, height=0.5)
ax.errorbar(diffs, return_df['Company'],
            xerr=[diffs - ci_lo_vals, ci_hi_vals - diffs],
            fmt='none', color='black', capsize=5, linewidth=1.5)
ax.axvline(0, color='gray', linestyle='--', linewidth=1)
ax.set_xlabel('Mean Return Difference: Good − Bad Market (%)', fontsize=11)
ax.set_title(
    'T-Test Results: Daily Returns — Good vs. Bad Market Conditions\n'
    '(Error bars = 95% CI  |  Bonferroni correction applied)',
    fontsize=12, fontweight='bold')

for i, row in return_df.iterrows():
    p_val = row['p (Bonferroni)']
    sig = '***' if p_val == '<0.001' else '*'
    p_str = 'p<.001' if p_val == '<0.001' else 'p={}'.format(p_val)
    ax.text(ci_hi_vals[i] + 0.01, row['Company'],
            ' t={}, {} {}'.format(row['t-stat'], p_str, sig),
            va='center', fontsize=9)

ax.set_xlim(left=0, right=ci_hi_vals.max() + 0.25)
plt.tight_layout()
plt.show()
```

##### Returns interpretation
We **reject H₀** for all five companies. Every FAANG stock earns significantly higher returns during good market conditions, even after Bonferroni correction. **Netflix** shows the largest gap between regimes (+1.13% per day), while **Google** and **Apple** show the smallest (~+0.66%), suggesting these two are comparatively more resilient. The confidence intervals for all companies are entirely above zero, confirming the direction of the effect is consistent.


##### Part 2: Annualized Volatility — Good vs. Bad Market Conditions

We repeat the same t-test framework on annualized rolling volatility (30-day window, scaled by √252). Higher volatility in bad markets is expected, but the **magnitude of the difference** across companies speaks directly to their relative resilience.

- **H₀**: Mean annualized volatility is equal across market regimes
- **H₁**: Mean annualized volatility is higher during bad market conditions

```python
vol_results = []
for comp in companies:
    sub = faang_df[faang_df['Company'] == comp]
    good = sub[sub['Market_Regime'] == 'Good']['Volatility'].dropna()
    bad  = sub[sub['Market_Regime'] == 'Bad']['Volatility'].dropna()
    t, p = stats.ttest_ind(good, bad)
    diff = (bad.mean() - good.mean()) * 100
    se = np.sqrt(good.std()**2/len(good) + bad.std()**2/len(bad)) * 100
    ci_lo, ci_hi = diff - 1.96*se, diff + 1.96*se
    p_bonf = min(p * 5, 1.0)
    vol_results.append({
        'Company':        comp,
        'Mean Good (%)':  round(good.mean()*100, 3),
        'Mean Bad (%)':   round(bad.mean()*100, 3),
        'Increase (%)':   round(diff, 3),
        '95% CI':         '[{:.3f}, {:.3f}]'.format(ci_lo, ci_hi),
        't-stat':         round(t, 3),
        'p (Bonferroni)': '<0.001' if p_bonf < 0.001 else round(p_bonf, 4)
    })

vol_df = pd.DataFrame(vol_results)
display(vol_df.style.set_caption('Table 2: T-Test Results — Annualized Volatility (Good vs. Bad Market)'))
```

#### Part 2: Annualized Volatility — Good vs. Bad Market Conditions
We repeat the same t-test framework on annualized rolling volatility (30-day window, scaled by √252). Higher volatility in bad markets is expected, but the magnitude of the difference across companies speaks to their relative resilience.

H₀: Mean annualized volatility is equal across market regimes

H₁: Mean annualized volatility is higher during bad market conditions

```python
fig, ax = plt.subplots(figsize=(11, 5))
bar_colors = [COMPANY_COLORS[c] for c in vol_df['Company']]
diffs = vol_df['Increase (%)'].values
ci_lo_vals = np.array([float(s.strip('[]').split(',')[0]) for s in vol_df['95% CI']])
ci_hi_vals = np.array([float(s.strip('[]').split(',')[1]) for s in vol_df['95% CI']])

ax.barh(vol_df['Company'], diffs, color=bar_colors, alpha=0.85, height=0.5)
ax.errorbar(diffs, vol_df['Company'],
            xerr=[diffs - ci_lo_vals, ci_hi_vals - diffs],
            fmt='none', color='black', capsize=5, linewidth=1.5)
ax.axvline(0, color='gray', linestyle='--', linewidth=1)
ax.set_xlabel('Volatility Increase in Bad vs. Good Markets (%)', fontsize=11)
ax.set_title(
    'T-Test Results: Annualized Volatility — Good vs. Bad Market Conditions\n'
    '(Error bars = 95% CI  |  Bonferroni correction applied)',
    fontsize=12, fontweight='bold')

for i, row in vol_df.iterrows():
    p_val = row['p (Bonferroni)']
    sig = '***' if p_val == '<0.001' else '*'
    p_str = 'p<.001' if p_val == '<0.001' else 'p={}'.format(p_val)
    ax.text(ci_hi_vals[i] + 0.1, row['Company'],
            ' t={}, {} {}'.format(row['t-stat'], p_str, sig),
            va='center', fontsize=9)

ax.set_xlim(left=0, right=ci_hi_vals.max() + 3)
plt.tight_layout()
plt.show()
```

**Volatility interpretation**

We **reject H₀** for all five companies. Volatility is significantly higher in bad market conditions across every FAANG stock after Bonferroni correction. **Amazon** shows the largest absolute volatility spike (+12.64%), while **Facebook** shows the smallest (+3.76%) and is the only company significant at p < 0.05 rather than p < 0.001. **Google** has the lowest baseline volatility in good markets (20.84%), reinforcing its profile as the most stable FAANG stock.

Both t-tests converge on the same conclusion: **market conditions as measured by VIX regime have a significant and consistent effect on FAANG stock behavior.**

> **Limitation:** Bad market days (n=994) are far fewer than good market days (n=8,821), which widens confidence intervals for bad-market estimates. Future work could address this through bootstrapped confidence intervals or a matched-sample design to equalize group sizes.
